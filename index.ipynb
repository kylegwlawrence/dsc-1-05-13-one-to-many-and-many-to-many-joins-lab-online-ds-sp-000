```python
#import libraries
import pandas as pd
import sqlite3

#create connection
conn = sqlite3.connect('data.sqlite', detect_types=sqlite3.PARSE_COLNAMES)

#create cursor
cur = conn.cursor()
```

```python
#create function to handle query
def query(sql, head=0):
    cur.execute(sql)
    df = pd.DataFrame(cur.fetchall())
    df.columns = [i[0] for i in cur.description]
    if head == 1:
        return df.head()
    else:
        return df
```

## Employees and their Office (a One-to-One join)

Return a list of all of the employees with their first name, last name and the city and state of the office that they work out of (if they have one). Include all employees and order them by their first name, then their last name.

```python
#Your code here
query("""select employees.firstName, employees.lastName, offices.city, offices.state from employees
left join offices on employees.officeCode = offices.officeCode
order by employees.firstName ASC, employees.lastName ASC;""")
```

## Customers and their Orders (a One-to-Many join)

Return a list of all of the customers. For each customer return a record for each of their order numbers, order dates and statuses.

```python
# Your code here
query("""select cust.customerName, ord.orderDate, ord.status, ord.orderNumber from customers cust
left join orders ord on cust.customerNumber = ord.customerNumber
order by cust.customerName ASC, ord.orderDate ASC, ord.status, ord.orderNumber;""")
```

## Orders and their Order Details (another One-to-Many join)

Return a list of orders. For each order return a record for each order detail within the order.

```python
# Your code here
query("""select ordtl.* from orders ord
left join orderdetails ordtl on ord.orderNumber = ordtl.orderNumber
order by ordtl.orderNumber, ordtl.orderLineNumber""")
```

## Orders, Order details and Product Details (a Many-to-Many Join)

Return a list of the orders with the the order information, and all the product information for each product in the order. 

```python
# Your code here
query("""select ord.*, prod.* from orders ord
left join orderdetails ordtl on ordtl.orderNumber = ord.orderNumber
left join products prod on ordtl.productCode = prod.productCode
order by ord.orderNumber, ord.orderDate, ord.requiredDate, ord.shippedDate""")
```
